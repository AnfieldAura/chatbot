```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
def scrape(site):
    service = Service("chromedriver.exe")
    driver = webdriver.Chrome(service=service)
    driver.get(site)
    section = driver.find_element(By.XPATH, '//section')
    print(section.text)
    driver.quit()
    elements = driver.find_elements(By.XPATH, "//a[contains(@class, 'dropdown-toggle')]")
    texts = [element.text for element in elements if element.text.strip()]    
    print(texts)
```

```python
import re
```

```python
def clean(text):
    text=re.sub(r'\s+',' ',text)
    text=re.sub(r'[^\w\s]','',text)
    return text.lower().strip()
cleaned=[clean(text) for text in texts]
```

```python
from transformers import AutoTokenizer,AutoModel
import torch

model_name = "sentence-transformers/all-MiniLM-L6-v2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

tokens = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

with torch.no_grad():
    outputs = model(**tokens)

embeddings = outputs.last_hidden_state.mean(dim=1) 
print(embeddings.numpy())  
```
